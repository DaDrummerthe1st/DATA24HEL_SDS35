# Regressionsanalys med Seaborn

För att utföra regressionsanalys i Python kan vi använda Seaborns `lmplot()`-funktion.

```python
import seaborn as sns; sns.set_theme(font_scale=.75)
```

```python
df = sns.load_dataset("anscombe")

sns.lmplot(data=df, x="x", y="y", col="dataset", hue="dataset", col_wrap=2, height=4)
# skuggade området är konfidensintervallet, se dataset III med sin avvikande datapunkt. Med 95% sannolikhet ligger den faktiska linjen inom konfidensintervallet
```

En `lmplot()` (lm står för *linear model*) kombinerar ett punktdiagram med en regressionslinje. Det skuggade området är konfidensintervallet för regressionslinjen.

Vi kan också titta på datasetet `mpg` som beskriver bränsleeffektiviteten (*miles per gallon*) hos ett antal olika bilmodeller.

```python
df = sns.load_dataset("mpg")
# Miles per gallon
```

```python
# 1 mile per gallon is equal to 235.214583 liters per 100 km.
mpg = input("How many mpg?")
l_per_km = mpg * 235.214583
```

`mpg` är vår beroende variabel som vi vill undersöka med hjälp av de andra variablerna. Den beroende variabeln brukar sättas på y-axeln. Vi börjar med att kolla på hur `displacement`, det vill säga motorvolymen, påverkar bränsleeffektiviteten.

```python
sns.lmplot(data=df, x="displacement", y="mpg", line_kws={"color": "green"})
```

Vi kan också kolla hur `horsepower`-variabeln påverkar `mpg`-variabeln.

```python
sns.lmplot(data=df, x="horsepower", y="mpg", line_kws={"color": "green"})
```

Här ser vi att sambandet inte ser helt linjärt ut. Om vi sätter parametern `order` till 2 låter vi regressionslinjen böja på sig.

```python
sns.lmplot(data=df, x="horsepower", y="mpg", order=2, line_kws={"color": "green"}) # order = 2 -> linjen får böja sig (kvadrat)

# Regression ligger ofta till grund för prediktionen i maskininlärning
# ci - argumentet (confidense interval) med högre siffra (uppåt 100+) beskriver att stickprovet inte är tillräckligt stort
```

Den här linjen följer fördelningen bättre!
